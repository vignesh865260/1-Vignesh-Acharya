```python
#solve linear programming problem with two variables using simplex method.

from scipy.optimize import linprog
```

```python
obj=[-3,-2]
```

```python
lhs_ineq=[[1,1],[1,-1]]
```

```python
rhs_ineq=[4,2]
```

```python
bnd=[(0,float("inf")),(0,float("inf"))]
```

```python
opt=linprog(c=obj,A_ub=lhs_ineq,b_ub=rhs_ineq,bounds=bnd,method="revised simplex")
```
